```python
import string
data = '''Natural Language Processing (NLP) is a subfield of linguistics, computer science, and artificial
intelligence concerned with the interactions between computers and human language. Its used to
analyze text, allowing machines to understand, interpret, and manipulate human language. NLP has
many real-world applications, including machine translation, sentiment analysis, and chatbots.'''
```

```python
#step 1:Tokenization

tokens = data.split()
tokens
```

```python
#step 2: Lowercasing

tokens_lowercase = [word.lower() for word in tokens]
tokens_lowercase
```

```python
#step 3: Punctuation removal
no_punc = [word.strip(string.punctuation) for word in tokens_lowercase]
no_punc
```

```python
#step 4: Stop Word Removal

stop_word = ["the", "is","and", "a", "in", "for", "of", "to"]

clean_tokens = [word for word in no_punc if word not in stop_word]
clean_tokens
```

```python
#step 5: Stemming

def stemming(word):
  endings = ["ed", "ing", "s", "es", "tion", "tions"]
  for end in endings:
    if word.endswith(end):
      return word[:-len(end)]
  return word

tokens_stemmed = [stemming(word) for word in clean_tokens]
tokens_stemmed
```

```python
#bonus:Lemmatization
lemm_dict = {"is":"be",
        "are":"be",
        "was":"be",
        "were":"be",
        "machines":"machine",
        "interactions":"interact",
        "processing":"process"}

tokens_lemm = [lemm_dict[word] if word in lemm_dict else word for word in clean_tokens]
tokens_lemm
```
