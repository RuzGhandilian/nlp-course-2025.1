```python
import nltk
from nltk.tokenize import word_tokenize
import string
import pandas as pd
import re

nltk.download("punkt")
nltk.download("punkt_tab")

col_name = df.columns[1]

def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word not in string.punctuation]
    clean_text = " ".join(filtered_tokens)
    return tokens, filtered_tokens, clean_text

df["tokens"], df["filtered_tokens"], df["clean_text"] = zip(*df[col_name].map(preprocess))
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    tokens = text.split()
    return tokens

def build_vocabulary(docs):
    vocab = set()
    for doc in docs:
        tokens = preprocess(doc)
        vocab.update(tokens)
    return sorted(list(vocab))
```

```python
def doc_to_bow(doc, vocab):
    tokens = preprocess(doc)
    vector = [0] * len(vocab)
    for token in tokens:
        if token in vocab:
            idx = vocab.index(token)
            vector[idx] += 1
    return vector
```

```python
def bow_representation(train_docs, test_docs):
    vocab = build_vocabulary(train_docs)
    train_vectors = [doc_to_bow(doc, vocab) for doc in train_docs]
    test_vectors = [doc_to_bow(doc, vocab) for doc in test_docs]
    return vocab, train_vectors, test_vectors

```
