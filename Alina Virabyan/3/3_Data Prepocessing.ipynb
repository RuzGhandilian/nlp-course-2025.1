```python
def remove_username(text):
    username = re.sub(r'@\w+', ' ', str(text))
    return username
remove_username(text = df['text'])
```

```python
def remove_links(text):
    links = re.sub(r'https?://\S+|www\.\S+', ' ', str(text))
    return links
remove_links(text = df['text'])
```

```python
def not_ascii_symbols(text): 
    asci = r'[^\x00-\x7F]+'
    asc = re.sub(asci, ' ', str(text))
    return asc
not_ascii_symbols(text = df['text'])
```

```python
def remove_stop_words(text):
    stop_words = ['the', 'a', 'an', 'and', 'or', 'in', 'on', 'for', 'to', 'is', 'it', 'of', 'at']      
    words = str(text).split()                         
    filtered_words = [w for w in words if w.lower() not in stop_words]  
    return ' '.join(filtered_words)               
remove_stop_words(text = df['text'])
```

```python
def remove_digits(text):
    digit = re.sub(r'\d+', ' ', str(text))
    return digit
remove_digits(text = df['text'] )
```

```python
def remove_special_characters(text):
    return re.sub(r'[^\w\s]', '', str(text))
remove_special_characters(text = df['text'])
```
